```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df = pd.merge(clinical_df, mouse_df)
# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint and Convert to DataFrame
tumorResponse_df = df.groupby(['Drug', 'Timepoint']).mean().iloc[:, 0].to_frame().reset_index()

# Preview DataFrame
tumorResponse_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint and Convert to DataFrame
tumor_error_df = df.groupby(['Drug', 'Timepoint']).sem().iloc[:, 2].to_frame().reset_index()

# Preview DataFrame
tumor_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumorResponse_pivot_df = tumorResponse_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
# Preview that Reformatting worked
tumorResponse_pivot_df
```

```python
# Generate the Plot (with Error Bars)
markers=['x', '+', 'v', '.', 'o', '*', '1', '2', '3', '4']
i = 0
label = []
for drug in tumorResponse_pivot_df:
    label.append(drug)
    plt.plot(tumorResponse_pivot_df[drug], linestyle='dotted', label=label[i], marker=markers[i])
    i = i + 1
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volumn (mm3)')
plt.title('Tumor Response to Treatement')
plt.errorbar(x=tumorResponse_df["Timepoint"], y=tumorResponse_df["Tumor Volume (mm3)"], yerr=tumor_error_df["Tumor Volume (mm3)"], fmt='none')
# Save the Figure
plt.savefig('Tumor Response to Treatement')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint and convert to DataFrame
metaResponse_df = df.groupby(["Drug", "Timepoint"]).mean().iloc[:, 1].to_frame().reset_index()

# Preview DataFrame
metaResponse_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint and convert to DataFrame
meta_error_df = df.groupby(["Drug", "Timepoint"]).sem().iloc[:, 0].to_frame().reset_index()

# Preview DataFrame
meta_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metaResponse_pivot_df = metaResponse_df.reset_index().pivot(index="Timepoint", columns="Drug", values='Metastatic Sites')

# Preview that Reformatting worked
metaResponse_pivot_df
```

```python
# Generate the Plot (with Error Bars)
markers=['x', '+', 'v', '.', 'o', '*', '1', '2', '3', '4']
i = 0
label = []
for drug in metaResponse_pivot_df:
    label.append(drug)
    plt.plot(metaResponse_pivot_df[drug], linestyle='dotted', label=label[i], marker=markers[i])
    i = i + 1
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
plt.xlabel('Treatment Duration(Days)')
plt.ylabel('Met. Sites')
plt.title('Metastatic Spread During Treatment')
plt.errorbar(x=metaResponse_df["Timepoint"], y=metaResponse_df["Metastatic Sites"], yerr=meta_error_df["Metastatic Sites"], fmt='none')

# Save the Figure
plt.savefig('Meta')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survRate_df = df.groupby(["Drug", "Timepoint"]).count().iloc[:, 0].to_frame().reset_index()
# Convert to DataFrame

# Preview DataFrame
survRate_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survRate_pivot_df = survRate_df.pivot(index='Timepoint', columns='Drug', values='Mouse ID')
# Preview the Data Frame
survRate_pivot_df
```

```python
# Store pectage info in a dataframe 
survPct_df = survRate_pivot_df*100/survRate_pivot_df.iloc[0, :]
```

```python
# Generate the Plot (Accounting for percentages)
markers=['x', '+', 'v', '.', 'o', '*', '1', '2', '3', '4']
i = 0
label = []
for drug in survPct_df:
    label.append(drug)
    plt.plot(survPct_df[drug], linestyle='dotted', label=label[i], marker=markers[i])
    i = i + 1
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
plt.xlabel('Time (Days)')
plt.ylabel('Surv Rate(%)')
plt.title('Survival During Treatment')

# Save the Figure
plt.savefig('Surv')
# Show the Figure
plt.show()
```
